```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Aufgabe 1**

```python
# Daten erzeugen
x = np.linspace(0, 40, 1000)  # x-Werte von 0 bis 10 mit 100 Punkten
y = np.sin(x) * np.exp(-0.2*x)  # Funktion: gedämpfte Sinusschwingung
#plt.style.use('default')
# Plot erstellen
plt.plot(x, y, color="darkred", linestyle="-.", linewidth=1, label="= ?")

# Achsenbeschriftungen und Titel
plt.xlabel("x-Werte")
plt.ylabel("y-Werte")
plt.title("Gedämpfte Sinusschwingung")

# Legende hinzufügen
plt.legend()
plt.grid(True)

# Plot anzeigen
plt.savefig("Quizz,matplotlib.png")
plt.show()
```

**Aufgabe 2**

```python
# Zufällige Daten erzeugen


#plt.style.use('default')
x =  np.random.uniform(0, 10, 50),  # 50 zufällige x-Werte zwischen 0 und 10
y = np.random.uniform(0, 10, 50)   # 50 zufällige y-Werte zwischen 0 und 10
# Scatter Plot erstellen
plt.figure(figsize=(7,5))
plt.scatter(x, y, color="blue", marker="x", s=50, label="Datenpunkte") 

# Achsenbeschriftungen und Titel
plt.xlabel("x-Werte")
plt.ylabel("y-Werte")
plt.title("Zufälliger Scatter Plot")

# Legende & Gitternetz
plt.legend()
plt.grid(True)

# Plot anzeigen
plt.show()
```

**Aufgabe 3**

```python
# DataFrame mit quadratischer Funktion erzeugen
x = np.linspace(0, 4.5, 100)  # 100 Werte von 0 bis 10
y = x**3 - 6*(x**2) + 9*x - 2

#Mehr Kosmetik, google oder chatgpt
plt.style.use('dark_background')
plt.gca().spines["top"].set_color("deeppink")
plt.gca().spines["bottom"].set_color("deeppink")
plt.gca().spines["left"].set_color("deeppink")
plt.gca().spines["right"].set_color("deeppink")

# Linienplot erstellen, in pink, in Farbe, mit label
plt.plot(x, y, color="deeppink", linewidth=5, marker="o", markerfacecolor="blue", label=r"$y = x^3 - 6x^2 +9x -2$")

# Achsenbeschriftungen und Titel
plt.xlabel("x-Werte", color = "deeppink") #label pink
plt.ylabel("y-Werte", color = "deeppink") #label pink
plt.tick_params(axis="x", colors="deeppink")  # X-Achse Zahlen pink
plt.tick_params(axis="y", colors="deeppink")  # Y-Achse Zahlen pink
plt.title("Quadratische Funktion", color = "deeppink")

# Legende & Gitternetz
plt.legend(edgecolor="deeppink", labelcolor="deeppink") #Rahmen pink, Beschreibung pink


# Plot anzeigen
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datensatz einlesen und Index setzen
climate = pd.read_csv("Climate.csv", sep="\t")  # Datensatz einlesen
clim = climate.set_index('time')  # Neues Datenset mit 'time' als Index

X = clim.index
Y = clim["summer"]
Y2 = clim["winter"]
plt.plot(X,Y, label= "summer")
plt.plot(X,Y2, label= "winter")
plt.legend()
```
